```python
import argparse
import itertools
import pickle
from perses.app.relative_point_mutation_setup import PointMutationExecutor
import simtk.openmm.app as app
import os

```

```python
amino_acids = ['ARG', 'ASN', 'ASP', 'CYS', 'GLN', 'GLU', 'GLY', 'HIS', 'ILE', 'LEU', 'LYS', 'MET', 'PHE',
                        'SER', 'THR', 'TRP', 'TYR', 'VAL']
```

```python
# Vacuum
wt = 'ALA'
for i, proposed in enumerate(amino_acids):
    print(f"vacuum: {wt}, {proposed}")

    # Create hybrid topology factory
    apo_delivery = PointMutationExecutor(f"../input/{wt.lower()}_vacuum.pdb", 
                              '1', 
                              '2', 
                              proposed,
                              phase='vacuum',
	                          barostat=None,
	                          periodic_forcefield_kwargs=None, 
	                          nonperiodic_forcefield_kwargs={'nonbondedMethod': app.NoCutoff}
                             )
    pickle.dump(apo_delivery.get_apo_htf(), open(os.path.join(outdir, f"{wt.lower()}_{proposed.lower()}_vacuum.pickle"), "wb" ))


```

```python
import mdtraj as md
import numpy as np
# save all pdbs in input
wt = 'ALA'
for proposed in amino_acids:
    print(proposed)
    htf = pickle.load(open(os.path.join(outdir, f"{wt.lower()}_{proposed.lower()}_vacuum.pickle"), "rb" ))
    new_top = md.Topology.from_openmm(htf._topology_proposal.new_topology)
    new_pos = htf.new_positions(htf.hybrid_positions)
    traj = md.Trajectory(np.array(new_pos), new_top)
    traj.save(os.path.join("../input/", f"{proposed.lower()}_vacuum.pdb"))
```

```python
import argparse
import itertools
import pickle
from perses.app.relative_point_mutation_setup import PointMutationExecutor
import simtk.openmm.app as app
import os
from perses.tests.test_topology_proposal import generate_atp, generate_dipeptide_top_pos_sys
```

```python
def render_protein_residue_atom_mapping(topology_proposal, filename):
    """
    wrap the `render_atom_mapping` method around protein point mutation topologies.
    TODO : make modification to `render_atom_mapping` so that the backbone atoms are not written in the output.
    
    arguments
        topology_proposal : perses.rjmc.topology_proposal.TopologyProposal object
            topology proposal of protein mutation
        filename : str
            filename to write the map
    """
    from perses.utils.smallmolecules import render_atom_mapping
    oe_res_maps = {}
    for omm_new_idx, omm_old_idx in topology_proposal._new_to_old_atom_map.items():
        if omm_new_idx in topology_proposal._new_topology.residue_to_oemol_map.keys():
            try:
                oe_res_maps[topology_proposal._new_topology.residue_to_oemol_map[omm_new_idx]] = topology_proposal._old_topology.residue_to_oemol_map[omm_old_idx]
            except:
                pass
            
    render_atom_mapping(filename, topology_proposal._old_topology.residue_oemol, topology_proposal._new_topology.residue_oemol, oe_res_maps)
    
```

```python
amino_acids = ['ARG', 'ASN', 'ASP', 'CYS', 'GLN', 'GLU', 'GLY', 'HIS', 'ILE', 'LEU', 'LYS', 'MET', 'PHE',
                        'SER', 'THR', 'TRP', 'TYR', 'VAL']
ring_amino_acids = ['HIS', 'PHE', 'TRP', 'TYR']
```

```python
outdir = "/data/chodera/zhangi/perses_benchmark/test_maps/fixed/"
```

```python
wt = 'ALA'
atp_test, system_generator = generate_atp(phase='vacuum')
d_top_pos_sys = {}
```

```python
for i, proposed in enumerate(amino_acids):
    print(f"vacuum: {wt}, {proposed}")
    
    validate_bool = False if proposed in ring_amino_acids else True
    htf = generate_dipeptide_top_pos_sys(atp_test.topology, proposed, atp_test.system, atp_test.positions, system_generator, conduct_geometry_prop=True, conduct_htf_prop=True, validate_energy_bookkeeping=validate_bool)
    d_top_pos_sys[proposed] = [htf._topology_proposal._new_topology, htf._new_system, htf._new_positions]
    
    render_protein_residue_atom_mapping(htf._topology_proposal, os.path.join(outdir, f"{wt.lower()}_{proposed.lower()}_vacuum.png"))

```

```python
import pickle
# save  top, pos, sys dict
pickle.dump(d_top_pos_sys, open(os.path.join(outdir, "d_top_pos_sys_ala.pickle"), "wb" ))
```
