# Forward neq (image complex traj)

```python
from simtk import openmm, unit
```

```python
with open("/data/chodera/zhangi/perses_benchmark/neq/6/13/13_complex.pickle", 'rb') as f:
    htf = pickle.load(f)
```

```python
vec = htf._topology_proposal.old_system.getDefaultPeriodicBoxVectors()
```

```python
vec
```

```python
vec = [val.value_in_unit_system(unit.md_unit_system) for val in vec]
```

```python
vecs = np.full((200, 3, 3), vec)
```

```python
vecs.shape
```

```python
with open(f"/data/chodera/zhangi/perses_benchmark/neq/6/13/13_complex_0_forward_neq_old.npy", 'rb') as f:
    old_pos = np.load(f)
```

```python
md_topology = md.Topology.from_openmm(htf._topology_proposal.old_topology)
```

```python
old_traj = md.Trajectory(old_pos, md_topology)
```

```python
old_traj.unitcell_vectors = vecs
```

```python
old_traj
```

```python
for residue in old_traj.topology.residues:
    print(residue.chain.index, residue)
```

```python
old_traj = old_traj.image_molecules(anchor_molecules=md_topology.find_molecules()[0:1])
```

```python
old_traj.save_dcd("/data/chodera/zhangi/perses_benchmark/neq/6/13/forward_neq/complex_old.dcd")
```
