```python
import pickle
import os
from perses.app.relative_point_mutation_setup import PointMutationExecutor
from simtk import unit
from simtk.openmm import app
```

# ALA -> ASP solvent with counterion fix (16/4)

```python
solvent_delivery = PointMutationExecutor("../../input/ala_vacuum.pdb",
                        '1',
                        '2',
                        'ASP',
                        ionic_strength=0.15*unit.molar,
                        flatten_torsions=True,
                        flatten_exceptions=True, 
                        conduct_endstate_validation=False, 
                        repartitioned_endstate=0
                       )

```

```python
outdir = "/data/chodera/zhangi/perses_benchmark/neq/16/4/"
apo_htf = solvent_delivery.get_apo_htf()
with open(os.path.join(outdir, f"4_solvent_0.pickle"), "wb") as f:
    pickle.dump(apo_htf, f)
```

```python
outdir = "/data/chodera/zhangi/perses_benchmark/neq/16/4/"
with open(os.path.join(outdir, f"4_solvent_0.pickle"), "rb") as f:
    htf = pickle.load(f)
```

```python
nb_force = htf._topology_proposal.new_system.getForce(3)
```

```python
for i in range(nb_force.getNumParticles()):
    charge, sigma, eps = nb_force.getParticleParameters(i)
    print(i, charge, sigma, eps)
```

```python
nb_force = htf.hybrid_system.getForce(4)
```

```python
for i in range(nb_force.getNumParticles()):
    charge, sigma, eps = nb_force.getParticleParameters(i)
    print(i, charge, sigma, eps)
```
