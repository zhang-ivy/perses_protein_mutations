# Make trajs 

```python
import mdtraj as md
import numpy as np
```

```python
outdir = "/data/chodera/zhangi/perses_benchmark/neq/16/10/"
with open(os.path.join(outdir, f"10_solvent.pickle"), "rb") as f:
    htf = pickle.load(f)
```

```python
with open(os.path.join("/data/chodera/zhangi/perses_benchmark/neq/16/10/10_solvent_0_forward_neq_old.npy"), "rb") as f:
    pos_forward_old = np.load(f)
```

```python
with open(os.path.join("/data/chodera/zhangi/perses_benchmark/neq/16/10/10_solvent_0_forward_neq_new.npy"), "rb") as f:
    pos_forward_new = np.load(f)
```

```python
with open(os.path.join("/data/chodera/zhangi/perses_benchmark/neq/16/10/10_solvent_0_reverse_neq_old.npy"), "rb") as f:
    pos_reverse_old = np.load(f)
```

```python
with open(os.path.join("/data/chodera/zhangi/perses_benchmark/neq/16/10/10_solvent_0_reverse_neq_new.npy"), "rb") as f:
    pos_reverse_new = np.load(f)
```

```python
forward_old = md.Trajectory(pos_forward_old, md.Topology.from_openmm(htf._topology_proposal.old_topology))
```

```python
forward_new = md.Trajectory(pos_forward_new, md.Topology.from_openmm(htf._topology_proposal.new_topology))
```

```python
reverse_old = md.Trajectory(pos_reverse_old, md.Topology.from_openmm(htf._topology_proposal.old_topology))
```

```python
reverse_new = md.Trajectory(pos_reverse_new, md.Topology.from_openmm(htf._topology_proposal.new_topology))
```

```python
forward_old.save(os.path.join(outdir, "10_solvent_0_forward_neq_old.dcd"))
```

```python
forward_old[0].save(os.path.join(outdir, "10_solvent_0_forward_neq_old.pdb"))
```

```python
forward_new.save(os.path.join(outdir, "10_solvent_0_forward_neq_new.dcd"))
```

```python
forward_new[0].save(os.path.join(outdir, "10_solvent_0_forward_neq_new.pdb"))
```

```python
reverse_old.save(os.path.join(outdir, "10_solvent_0_reverse_neq_old.dcd"))
```

```python
reverse_new.save(os.path.join(outdir, "10_solvent_0_reverse_neq_new.dcd"))
```
