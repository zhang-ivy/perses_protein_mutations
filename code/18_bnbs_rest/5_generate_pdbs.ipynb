Generate mutated PDBs for reverse transformations

### D35A barstar

```python
from simtk import openmm, unit
from simtk.openmm import app
import numpy as np
import pickle
import os
from perses.app.relative_point_mutation_setup import PointMutationExecutor
import mdtraj as md
```

```python
# Set up logger
import logging
_logger = logging.getLogger()
_logger.setLevel(logging.INFO)
```

```python
# Create htf
delivery = PointMutationExecutor("../../input/mmc2_barstar.pdb",
                        '1', # First chain is the barstar one
                        '35',
                        'ALA',
                        phase='vacuum',
                        barostat=None,
                         periodic_forcefield_kwargs=None,
                         nonperiodic_forcefield_kwargs={'nonbondedMethod': app.NoCutoff},
                        flatten_torsions=True,
                        flatten_exceptions=True      
                       )
```

```python
htf = delivery.get_apo_htf()
```

```python
out_dir = '../../input/'
traj = md.Trajectory(np.array(htf.new_positions(htf.hybrid_positions)), md.Topology.from_openmm(htf._topology_proposal.new_topology))
traj[0].save_pdb(os.path.join(out_dir, "mmc2_barstar_D35A.pdb"))
```
