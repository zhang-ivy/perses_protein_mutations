## Fix residue numbering for NME and all residues after

```python
from simtk.openmm import app
```

```python
pdb = app.PDBFile("../../input/1brs_barstar.pdb")
```

```python
for residue in pdb.topology.residues():
    if residue.name == 'NME' or residue.name == 'HOH':
        residue.id = str(int(residue.id) + 1)
        residue.insertionCode = ' '
```

```python
app.PDBFile.writeFile(pdb.topology, pdb.positions, open("../../input/1brs_barstar_renumbered.pdb", "w"), keepIds=True)
```

## Make barnase PDB for LYN27

```python
from simtk.openmm import app, unit
```

```python
pdb = app.PDBFile("../../input/1brs_barnase_renumbered.pdb")
```

```python
# Create list of variants to use for each residue. None means the variant will be selected automatically
variants = []
for res in pdb.topology.residues():
    if res.id == '27' and res.name == 'LYS':
        variants.append('LYN')
    else:
        variants.append(None)
```

```python
# Call addHydrogens() with the above variants list to protonate LYS27 appropriately (s.t. its LYN)
modeller = app.Modeller(pdb.topology, pdb.positions)
modeller.addHydrogens(pH=8.0, variants=variants)
```

```python
# ff = app.ForceField(*['amber14/protein.ff14SB.xml', 'amber14/tip3p.xml'])
# modeller.addSolvent(ff, padding=0.9 * unit.nanometers, ionicStrength=0.05 * unit.molar)
```

```python
app.PDBFile.writeFile(modeller.topology, modeller.positions, open("../../input/1brs_barnase_renumbered_lyn27.pdb", "w"), keepIds=True)
```
