```python
import mdtraj as md
import pickle
import numpy as np
from matplotlib import pyplot as plt
from simtk import openmm, unit
from tqdm import tqdm

import logging
# Set up logger
_logger = logging.getLogger()
_logger.setLevel(logging.INFO)
```

### 20 ns of equilibration for ALA (vanilla system)

```python
with open("/data/chodera/zhangi/perses_benchmark/neq/7/25/25_vacuum.pickle", 'rb') as f:
    htf = pickle.load(f)
```

```python
old_top = md.Topology.from_openmm(htf._topology_proposal.old_topology)
```

```python
with open("/data/chodera/zhangi/perses_benchmark/neq/7/25/0_vacuum_positions_20ns.npy", 'rb') as f:
    pos = np.load(f)
```

```python
traj = md.Trajectory(np.array(pos), old_top)
```

```python
for atom in traj.topology.atoms:
    print(atom, atom.index)
```

```python
dihedrals = np.sin(md.compute_dihedrals(traj, np.array([[15, 10, 8, 6]])))
```

```python
plt.hist(dihedrals)
plt.show()
```

```python
dihedrals_notsin = md.compute_dihedrals(traj, np.array([[15, 10, 8, 6]]))
```

```python
plt.hist(dihedrals_notsin)
plt.show()
```

```python
plt.scatter(range(4000), dihedrals)
plt.ylabel("sin(dihedral)")
plt.xlabel("iteration number")
plt.show()
```

```python
from perses.dispersed import feptasks
```

```python
t0, g, neff_max, a_t, uncorrelated_indices = feptasks.compute_timeseries(dihedrals)
```

```python
plt.hist(dihedrals[uncorrelated_indices])
plt.show()
```
