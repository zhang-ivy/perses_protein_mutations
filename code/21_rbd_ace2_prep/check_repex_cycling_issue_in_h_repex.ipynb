```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
with open("/data/chodera/zhangi/perses_benchmark/repex/31/1/0/log.txt", "r") as f:
    lines = f.readlines()
```

```python
times = []
for line in lines:
    if "Iteration took" in line:
        time = float(line.split(" ")[-1].strip('s.\n'))
        times.append(time)
```

```python
plt.plot(times)
plt.xlabel('iteration number')
plt.ylabel('iteration length (s)')
plt.title('N501Y complex h-repex')
```
