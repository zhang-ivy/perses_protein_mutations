```python
import pickle
import os
from perses.app.relative_point_mutation_setup import PointMutationExecutor
from simtk import unit
from simtk.openmm import app
```

```python
solvent_delivery = PointMutationExecutor("../../input/asn_vacuum.pdb",
                        '1',
                        '2',
                        'LYS',
                        ionic_strength=0.15*unit.molar,
                        flatten_torsions=True,
                        flatten_exceptions=True, 
                        conduct_endstate_validation=False
                       )

```

```python
outdir = "./"
apo_htf = solvent_delivery.get_apo_htf()
with open(os.path.join(outdir, f"ASN_LYS.pickle"), "wb") as f:
    pickle.dump(apo_htf, f)
```

```python
def write(path, name):
    i = os.path.basename(os.path.dirname(outdir))
    with open(path, "rb") as f:
        htf = pickle.load(f)
    
    # Render atom map
    atom_map_filename = f'{outdir}/{name}.png'
    render_protein_residue_atom_mapping(htf._topology_proposal, atom_map_filename)
```
