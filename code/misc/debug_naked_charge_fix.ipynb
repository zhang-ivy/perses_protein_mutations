```python
from perses.app.relative_point_mutation_setup import PointMutationExecutor
from pkg_resources import resource_filename
from simtk.openmm import unit, app
```

```python
pdb_filename = resource_filename("perses", "data/ala_vacuum.pdb")
solvent_delivery = PointMutationExecutor(
    pdb_filename,
    "1",
    "2",
    "ASP",
    flatten_torsions=True,
    flatten_exceptions=True,
    conduct_endstate_validation=False,
    barostat=None,
    phase="vaccum",
    periodic_forcefield_kwargs=None,
    nonperiodic_forcefield_kwargs={"nonbondedMethod": app.NoCutoff},
)
htf = solvent_delivery.get_apo_htf()
```

```python
old_system = htf._topology_proposal.old_system
new_system = htf._topology_proposal.new_system
old_topology_res_map = {atom.index: atom.residue.name for atom in htf._topology_proposal.old_topology.atoms()}
new_topology_res_map = {atom.index: atom.residue.name for atom in htf._topology_proposal.new_topology.atoms()}

for i, system in enumerate([htf._topology_proposal.old_system, htf._topology_proposal.new_system]):
    force_dict = {i.__class__.__name__: i for i in old_system.getForces()}
    res_map = old_topology_res_map if i == 0 else new_topology_res_map
    print(i)
    if 'NonbondedForce' in [i for i in force_dict.keys()]:
        nb_force = force_dict['NonbondedForce']
        for idx in range(nb_force.getNumParticles()):
            if res_map[idx] in ['HOH', 'WAT']:
                continue
            charge, sigma, epsilon = nb_force.getParticleParameters(idx)
    #         print(i, charge, sigma, epsilon)
            if sigma == 0*unit.nanometer:
                print(f"setting particle {idx}'s sigma from {sigma} to 0.06")
    #             sigma = 0.06*unit.nanometer
    #             nb_force.setParticleParameters(i, charge, sigma, epsilon)
            if epsilon == 0*unit.kilojoule_per_mole:
                print(f"setting particle {idx}'s epsilon from {epsilon} to 0.0001")
    #             epsilon = 0.0001*unit.kilojoule_per_mole
    #             nb_force.setParticleParameters(i, charge, sigma, epsilon)
```
