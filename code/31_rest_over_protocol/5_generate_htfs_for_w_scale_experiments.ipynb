## Generate htfs for w_scale = 0.001, 0.01, 0.1, 0.3, 0.4, 1.0

```python
# Create a htf
atp, system_generator = generate_atp(phase = 'solvent')
topology_proposal, new_positions, _, _ = generate_dipeptide_top_pos_sys(atp.topology, 
                               'THR', 
                               atp.system, 
                               atp.positions, 
                               system_generator)
```

```python
w_scales = [0.001, 0.01, 0.1, 0.3, 0.4, 1.0]
for i, w_scale in enumerate(w_scales):
    subdir = 3 + i
    print("GENERATING HTF WITH W_SCALE: ", w_scale)
    htf = RESTCapableHybridTopologyFactory(topology_proposal=topology_proposal,
                         current_positions=atp.positions,
                         new_positions=new_positions,
                         w_scale=w_scale
                        )
    with open(f"/data/chodera/zhangi/perses_benchmark/repex/32/3/{subdir}/{subdir}_apo.pickle", "wb") as f:
        pickle.dump(htf, f)
```

### Get code for turning off unique old atoms in a force

```python
electrostatics_force = htf.hybrid_system.getForce(4)
for i in range(electrostatics_force.getNumParticles()):
    params = electrostatics_force.getParticleParameters(i)
    if params[5] == 1:
        params = list(params)
        params[-2] = params[-2]*0
        electrostatics_force.setParticleParameters(i, params)
        print(i, params)
```

```python
sterics_force = htf.hybrid_system.getForce(5)
for i in range(sterics_force.getNumParticles()):
    params = sterics_force.getParticleParameters(i)
    if params[5] == 1:
        params = list(params)
        params[-2] = params[-2]*0
        sterics_force.setParticleParameters(i, params)
        print(i, params)
```

```python
reciprocal_force = htf.hybrid_system.getForce(7)
for i in range(reciprocal_force.getNumParticles()):
    charge, sigma, epsilon = reciprocal_force.getParticleParameters(i)
    if i in htf._atom_classes['unique_old_atoms']:
        reciprocal_force.setParticleParameters(i, charge*0, sigma, epsilon)
        print(i)
```
