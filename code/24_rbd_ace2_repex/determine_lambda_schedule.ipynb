```python
import numpy as np
```

```python
n_states = 24
```

```python
# default protocol
for i in np.linspace(0.,1.,n_states):
    if i <= 0.5:
        print(f"0.0-0.5: {i}")
    else:
        print(f"0.5-1.0: {i}")
```

```python
# multiphase protocol
for index, i in enumerate(np.linspace(0.,1.,n_states)):
    if i <= 0.2:
        print(f"{index} 0.0-0.2: {i}")
    elif i > 0.2 and i <= 0.4:
        print(f"{index} 0.2-0.4: {i}")
    elif i > 0.4 and i <= 0.6:
        print(f"{index} 0.4-0.6: {i}")
    elif i > 0.6 and i <= 0.8:
        print(f"{index} 0.6-0.8: {i}")
    elif i > 0.8 and i <= 2:
        print(f"{index} 0.8-1.0: {i}")
```

```python
for i in np.linspace(0.,1.,n_states):
    if i <= 0.2:
        print(f"0.0-0.2: {i}")
    elif i > 0.2 and i <= 0.4:
        print(f"0.2-0.4: {i}")
    elif i > 0.4 and i <= 0.6:
        print(f"0.4-0.6: {i}")
    elif i > 0.6 and i <= 0.8:
        print(f"0.6-0.8: {i}")
    elif i > 0.8 and i <= 2:
        print(f"0.8-1.0: {i}")
```

```python
counter = 0
for i in np.linspace(0.0,0.4, 17):
    if i <= 0.2:
        print(f"{counter} 0.0-0.2: {i}")
    else:
        print(f"{counter} 0.2-0.4: {i}")
    counter += 1
print(f"{counter} 0.4-0.6: {0.5}")
counter += 1
for i in np.linspace(0.6,1.0, 17):
    if i <= 0.8:  
        print(f"{counter} 0.6-0.8: {i}")
    else:
        print(f"{counter} 0.8-1.0: {i}")
    counter += 1
```

# Multiphase protocol with more states in 0.7-0.8 and less in others

```python
counter = 0
for i in np.linspace(0.0,0.4, 14):
    if i <= 0.2:
        print(f"{counter} 0.0-0.2: {i}")
    else:
        print(f"{counter} 0.2-0.4: {i}")
    counter += 1
print(f"{counter} 0.4-0.6: {0.5}")
counter += 1
for i in np.linspace(0.6,0.7, 6):
    print(f"{counter} 0.6-0.7: {i}")
    counter += 1
for i in np.linspace(0.71, 0.8, 6): 
    print(f"{counter} 0.7-0.8: {i}")
    counter += 1
for i in np.linspace(0.81, 1.0, 7):
    print(f"{counter} 0.8-1.0: {i}")
    counter += 1
```
